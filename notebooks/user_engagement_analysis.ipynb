User Engagement analysis 

```python
import pandas as pd
import numpy as np
```

```python
cdf = pd.read_csv('../data/clean_TellCo_data.csv')
cdf.head()
```

Tracking user engagement using metrics:
* Session frequency: bearer id
* duration of the session : totla duratin of xDR
* session total traffic (DL and UL) : sum of Total DL and Total UL 

```python
# Adding two columns then creating new column and round floalt to 1 descimal places

pd.set_option('display.float_format', lambda x: '%.1f' % x)

cdf['Total_UL_and_DL'] = cdf['Total UL (Bytes)'] + cdf['Total DL (Bytes)']
cdf['Dur. (ms)'] = cdf['Dur. (ms)'] + cdf['Dur. (ms)']
```

```python
# Creating new data from the cdf

clean_column = ['Bearer Id','Dur. (ms)','MSISDN/Number','Total_UL_and_DL']
user_engagement_data = cdf[clean_column]
user_engagement_data.head()
```

```python
# Grouping Using MSISDN/Number
df = user_engagement_data.copy(deep= True)
aggregated_users = {'Bearer Id':'count','Dur. (ms)':'sum','Total_UL_and_DL':'sum'}

df = df.groupby(['MSISDN/Number']).agg(aggregated_users)
df
```
